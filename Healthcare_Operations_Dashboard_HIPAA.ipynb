#*HIPAA = Health Insurance Portability and Accountability*
HIPAA principles are the core rules that govern how healthcare data—especially patient data—must be handled, protected, and shared.

HIPAA = Health Insurance Portability and Accountability Act (USA)
Its purpose is to protect patient health information (PHI) from misuse, breaches, or unauthorized access.

*“HIPAA principles ensure that patient health data is accessed, used, and stored securely, only by authorized personnel, using the minimum necessary information.”*

PHI (Protected Health Information)?

PHI is any data that can identify a patient, such as:


*   Name, address, phone number

*   Medical record number

*   Dates (birth, admission, discharge)
*   Lab reports, scans, diagnoses


*   Any combination that identifies a person + health info

If data can identify a patient → HIPAA applies.



Core HIPAA Principles (Simple & Practical)
1. Privacy Rule – “Who can see the data?”

---> Only authorized people can access PHI

---> Access must be for a legitimate healthcare purpose

---> Use minimum necessary data

Example (Analyst):

It should not see patient names if age and diagnosis are enough.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Create a HIPAA-Compliant Dataset (Inside Colab)
data = {
    "Patient_ID": ["P001","P002","P003","P004","P005"],
    "Age": [45, 62, 30, 55, 40],
    "Gender": ["M","F","F","M","F"],
    "Department": ["Cardiology","Orthopedics","Neurology","Cardiology","General"],
    "Admission_Date": pd.to_datetime(["2025-01-01","2025-01-02","2025-01-03","2025-01-04","2025-01-05"]),
    "Discharge_Date": pd.to_datetime(["2025-01-05","2025-01-06","2025-01-04","2025-01-08","2025-01-07"]),
    "Admission_Type": ["Emergency","Scheduled","Emergency","Scheduled","Emergency"]
}

df = pd.DataFrame(data)
df["Length_of_Stay"] = (df["Discharge_Date"] - df["Admission_Date"]).dt.days
```

```python
df
```

*HIPAA Check*

No names

No phone numbers

No address

Uses Patient_ID only

##Healthcare Operations Analysis

```python
#Total Admissions
total_admissions = df.shape[0]
total_admissions
```

```python
# Admissions by Department
dept_counts = df['Department'].value_counts()
dept_counts
```

```python
# Average Lenght of Stay
avg_stay = df["Length_of_Stay"].mean()
avg_stay
```

```python
# Emergency vs Scheduled
admission_type_counts = df["Admission_Type"].value_counts()
admission_type_counts
```

#Visualization (Dashboard Elements)

```python
#Bar Chart – Patients by Department
dept_counts.plot(kind="bar")
plt.title("Patients by Department")
plt.xlabel("Department")
plt.ylabel("Number of Patients")
plt.show()
```

```python
#Pie Chart – Admission Type
admission_type_counts.plot(kind="pie", autopct="%1.1f%%")
plt.title("Admission Type Distribution")
plt.ylabel("")
plt.show()
```
